

### Theoretical Background

**Hoeffding's Inequality**: For bounded random variables $a_i \leq X_i \leq b_i$, the empirical mean $\overline{X} = \frac{1}{n}\sum_{i=1}^n X_i$ satisfies:
$$P(|\overline{X} - E[\overline{X}]| \geq \epsilon) \leq 2e^{-\frac{2n\epsilon^2}{(b-a)^2}}$$

**Chebyshev's Inequality**: For any random variable with finite variance:
$$P(|X - E[X]| \geq \epsilon) \leq \frac{Var(X)}{\epsilon^2}$$

## Theoretical Calculations for Confidence Intervals

### 1. Theoretical Formulas

#### Hoeffding's Inequality
For bounded random variables $a \leq X_i \leq b$, the confidence interval width is calculated as:

$$
\epsilon = \sqrt{\frac{(b-a)^2 \cdot \ln(2/\delta)}{2n}}
$$

Where:
- $\delta = 1 - \text{confidence level} = 0.05$ (for 95% confidence)
- $n = 1000$ (sample size)
- $b-a$ = range of the distribution

#### Chebyshev's Inequality
For any random variable with finite variance:

$$
\epsilon = \sqrt{\frac{\sigma^2}{\delta}}
$$

Where:
- $\sigma^2$ = theoretical variance
- $\delta = 0.05$

The confidence interval width in both cases is $2\epsilon$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns 

class DistributionAnalyzer:
    def __init__(self, sample_size=1000, num_experiments=100):
        self.sample_size = sample_size
        self.num_experiments = num_experiments
        self.results = {}
        
    def generate_distributions(self):
        # Generate random data from different distributions
        np.random.seed(42)  
        
        # (a) Normal distributions with different means and variances
        normal_data = {
            'normal_1': np.random.normal(0, 1, self.sample_size),      
            'normal_2': np.random.normal(5, 2.5, self.sample_size),    
            'normal_3': np.random.normal(-9, 3, self.sample_size)      
        }
        
        # (b) Uniform distribution
        uniform_data = np.random.uniform(0, 5, self.sample_size)
        
        # (c) Beta distributions with different parameters
        beta_data = {
            'beta_1': np.random.beta(2, 5, self.sample_size),    
            'beta_2': np.random.beta(0.5, 0.5, self.sample_size), 
            'beta_3': np.random.beta(2, 2, self.sample_size)     
        }
        
        return normal_data, uniform_data, beta_data
    
    def hoeffding_inequality(self, epsilon, range_val):
        
        n = self.sample_size
        confidence = 0.95
        delta = 1 - confidence
        
        # Solve: 2exp(-2nε²/(b-a)²) = delta
        epsilon_hoeffding = np.sqrt((range_val**2 * np.log(2/delta)) / (2 * n))
        
        return 2 * epsilon_hoeffding  # Return full interval width
    
    def chebyshev_inequality(self, variance, epsilon):
       
        confidence = 0.95
        delta = 1 - confidence
        
        # Solve: σ²/ε² = delta
        epsilon_chebyshev = np.sqrt(variance / delta)
        
        return 2 * epsilon_chebyshev  # Return full interval width
    
    def empirical_confidence_interval(self, data, confidence=0.95):
        
        mean = np.mean(data)
        std_err = stats.sem(data)  # Standard error
        ci = std_err * stats.t.ppf((1 + confidence) / 2, len(data)-1)
        return mean - ci, mean + ci, 2 * ci  # Return lower bound, upper bound, and width
    
    def analyze_distributions(self):
        
        normal_data, uniform_data, beta_data = self.generate_distributions()
        
        # Analyze normal distributions
        print(" Normal Distributions Analysis ")
        for name, data in normal_data.items():
            self._analyze_single_distribution(name, data, (-4, 4))
        
        # Analyze uniform distribution
        print("\nUniform Distribution Analysis ")
        self._analyze_single_distribution('uniform', uniform_data, (0, 1))
        
        # Analyze beta distributions
        print("\nBeta Distributions Analysis ")
        for name, data in beta_data.items():
            self._analyze_single_distribution(name, data, (0, 1))
    
    def _analyze_single_distribution(self, name, data, range_val):
        
        mean = np.mean(data)
        variance = np.var(data)
        range_width = range_val[1] - range_val[0]
        
        # Calculate various confidence intervals
        hoeffding_ci_width = self.hoeffding_inequality(0.05, range_width)
        chebyshev_ci_width = self.chebyshev_inequality(variance, 0.05)
        empirical_lower, empirical_upper, empirical_width = self.empirical_confidence_interval(data)
        
        print(f"\n{name}:")
        print(f"  Sample mean: {mean:.4f}")
        print(f"  Sample variance: {variance:.4f}")
        print(f"  Hoeffding Confidence interval width: {hoeffding_ci_width:.4f}")
        print(f"  Chebyshev Confidence interval width: {chebyshev_ci_width:.4f}")
        print(f"  Empirical Confidence interval width: {empirical_width:.4f}")
        print(f"  Empirical Confidence interval: [{empirical_lower:.4f}, {empirical_upper:.4f}]")
        
        # Store results
        self.results[name] = {
            'mean': mean,
            'variance': variance,
            'hoeffding_width': hoeffding_ci_width,
            'chebyshev_width': chebyshev_ci_width,
            'empirical_width': empirical_width,
            'data': data
        }
    
    def plot_essential_comparisons(self):
        
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))
        
        #  Confidence interval width comparison
        ax1 = axes[0]
        names = list(self.results.keys())
        hoeffding_widths = [self.results[name]['hoeffding_width'] for name in names]
        chebyshev_widths = [self.results[name]['chebyshev_width'] for name in names]
        empirical_widths = [self.results[name]['empirical_width'] for name in names]
        
        x = np.arange(len(names))
        width = 0.25
        
        ax1.bar(x - width, hoeffding_widths, width, label='Hoeffding', alpha=0.7)
        ax1.bar(x, chebyshev_widths, width, label='Chebyshev', alpha=0.7)
        ax1.bar(x + width, empirical_widths, width, label='Empirical', alpha=0.7)
        
        ax1.set_xlabel('Distribution')
        ax1.set_ylabel('Confidence Interval Width')
        ax1.set_title('Confidence Interval Width Comparison (Problem 1)')
        ax1.set_xticks(x)
        ax1.set_xticklabels(names, rotation=45)
        ax1.legend()
        
        #  Check 68-95-99.7 rule for normal distributions
        ax2 = axes[1]
        normal_names = [name for name in names if 'normal' in name]
        
        # Theoretical percentages for normal distribution
        theoretical_percentages = [68.27, 95.45, 99.73]
        
        # Calculate empirical percentages for each normal distribution
        for i, name in enumerate(normal_names):
            data = self.results[name]['data']
            mean = np.mean(data)
            std = np.std(data)
            
            empirical_percentages = [
                np.mean((data >= mean - 1*std) & (data <= mean + 1*std)) * 100,
                np.mean((data >= mean - 2*std) & (data <= mean + 2*std)) * 100,
                np.mean((data >= mean - 3*std) & (data <= mean + 3*std)) * 100
            ]
            
            x_pos = np.arange(3) + i * 0.25
            ax2.bar(x_pos, empirical_percentages, width=0.2, 
                   label=f'{name} (empirical)', alpha=0.7)
        
        # Add theoretical line
        ax2.plot([-0.2, 2.7], [theoretical_percentages[0], theoretical_percentages[0]], 
                'k--', label='Theoretical 68.27%', alpha=0.7)
        ax2.plot([-0.2, 2.7], [theoretical_percentages[1], theoretical_percentages[1]], 
                'k:', label='Theoretical 95.45%', alpha=0.7)
        ax2.plot([-0.2, 2.7], [theoretical_percentages[2], theoretical_percentages[2]], 
                'k-.', label='Theoretical 99.73%', alpha=0.7)
        
        ax2.set_xlabel('Standard Deviations')
        ax2.set_ylabel('Percentage of Data (%)')
        ax2.set_title('68-95-99.7 Rule Check (Problem 3)')
        ax2.set_xticks([0.25, 1.25, 2.25])
        ax2.set_xticklabels(['±1σ', '±2σ', '±3σ'])
        ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax2.set_ylim(0, 100)
        
        plt.tight_layout()
        plt.show()
    
    def check_68_95_99_7_rule(self):
        #Check which distributions follow the 68-95-99.7 rule
        print("\n" + "="*60)
        print("68-95-99.7 Rule Analysis (Problem 3)")
        print("="*60)
        
        for name, result in self.results.items():
            data = result['data']
            mean = np.mean(data)
            std = np.std(data)
            
            within_1std = np.mean((data >= mean - 1*std) & (data <= mean + 1*std)) * 100
            within_2std = np.mean((data >= mean - 2*std) & (data <= mean + 2*std)) * 100
            within_3std = np.mean((data >= mean - 3*std) & (data <= mean + 3*std)) * 100
            
            print(f"\n{name}:")
            print(f"  Within ±1σ: {within_1std:.2f}% (theoretical: 68.27%)")
            print(f"  Within ±2σ: {within_2std:.2f}% (theoretical: 95.45%)")
            print(f"  Within ±3σ: {within_3std:.2f}% (theoretical: 99.73%)")
            
            # Check if distribution follows the rule
            follows_rule = (abs(within_1std - 68.27) < 5 and 
                          abs(within_2std - 95.45) < 3 and 
                          abs(within_3std - 99.73) < 1)
            
            if follows_rule:
                print(f"  FOLLOWS the 68-95-99.7 rule")
            else:
                print(f"  DOES NOT follow the 68-95-99.7 rule")
    
    def run_analysis(self):
        
        
       
        
        self.analyze_distributions()
        self.plot_essential_comparisons()
        self.check_68_95_99_7_rule()
        
      
    
    
       

# Run analysis
if __name__ == "__main__":
    analyzer = DistributionAnalyzer(sample_size=1000, num_experiments=100)
    analyzer.run_analysis()
```
